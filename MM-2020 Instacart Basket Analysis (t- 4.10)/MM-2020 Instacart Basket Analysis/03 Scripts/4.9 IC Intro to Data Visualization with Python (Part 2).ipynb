## Task Step 3- Provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column

```python
hist = ords_prods_cust_merged['order_hour_of_day'].plot.hist(bins = 24)
```

The histogram shows spikes at specific hours, revealing when the most orders are placed. If the peak happens around midday or early evening, it suggests that customers tend to shop during lunch breaks or after work. The distribution's shape can indicate whether ordering patterns are clustered at certain times or spread out evenly throughout the day. A normal distribution would imply a steady shopping pattern around a central time, whereas a skewed distribution would point to a preference for shopping at particular times, such as in the afternoon or evening.

## Task Step 5- Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day

```python
# Create a random sample by generating a seed
np.random.seed(4)
```

```python
# Create a dev list based on the data split we want (70/30)
dev = np.random.rand(len(ords_prods_cust_merged)) <= 0.7
```

```python
# Split the dataframe into two samples
big = ords_prods_cust_merged[dev]
small = ords_prods_cust_merged[~dev]
```

```python
# Compare the dataframe length to the sum of the big and small samples
len(ords_prods_cust_merged)
```

```python
len(big) + len(small)
```

```python
# Reduce the samples to only the columns needed
ords_prods_cust_merged_2 = small [['order_hour_of_day','prices']]
```

```python
# Check if the subset of the sample was made properly
ords_prods_cust_merged_2.head(5)
```

```python
# Create a line chart
line = sns.lineplot(data = ords_prods_cust_merged_2, x = 'order_hour_of_day',y = 'prices')
```

## Task Step 6- Determine whether there’s a connection between age and family situation by creating a line chart exploring the connections between age and number of dependents

```python
# Reduce previously made sample to only the columns needed
ords_prods_cust_merged_3 = small [['age','n_dependants']]
```

```python
# Check if the subset of the sample was made properly
ords_prods_cust_merged_3.head(5)
```

## Task Step 7- Explore whether there’s a connection between age and spending power (income) using a scatterplot

```python
# Reduce previously made sample to only the columns needed
ords_prods_cust_merged_4 = small [['age','income']]
```

```python
# Create a scatterplot
scatterplot = sns.scatterplot(x = 'age', y = 'income',data = ords_prods_cust_merged_4)
```

Among younger age groups, particularly those between 20 and 40, incomes vary widely, with a notable concentration at lower levels. However, as individuals reach the age of 40 and beyond, there seems to be a notable upward trajectory in income, with many falling into higher income brackets. This suggests a potential correlation between age and income increase. The data points for individuals over 40 are noticeably clustered at higher income levels compared to those under 40, indicating a possible rise in median income with age. While the scatterplot visually suggests these trends, further statistical analysis would be needed to confirm any correlations or understand the intricacies of the relationship between age and income among the demographic.
